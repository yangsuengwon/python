```python
import csv
import matplotlib.pyplot as plt
n=list(range(1,13))
f=open('c:/201930221/weather.csv')
data=csv.reader(f)
header=next(data)
monthly_wind=[0 for x in range(12)]
day=[0 for x in range(12)]
for row in data:
    month=int(row[0][5:7])
    if row[3] !='':
        wind=float(row[3])
        monthly_wind[month-1]+=wind
        day[month-1]+=1
for i in range(12):
    monthly_wind[i]/=day[i]
plt.plot(monthly_wind,'blue')
plt.xticks(list(range(0,12)),n)
plt.show()
f.close()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
iris=load_iris()
num_neigh=5
knn=KNeighborsClassifier(n_neighbors=num_neigh)
knn.fit(iris.data,iris.target)
y_pred_all=knn.predict(iris.data)
scores=metrics.accuracy_score(iris.target,y_pred_all)
print('n_neighbors가 {0:d}일때 정확도: {1:.3f}'.format(num_neigh,scores))
plt.hist2d(iris.target,y_pred_all,bins=(3,3),cmap=plt.cm.jet)
conf_mat=confusion_matrix(iris.target,y_pred_all)
print(conf_mat)
plt.matshow(conf_mat,cmap=plt.cm.gray)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
life=pd.read_csv('c:/201930221/life_expectancy.csv')
life.dropna(inplace=True)
X=life[['Alcohol','Percentage expenditure','Polio','BMI','GDP','Thinness 1-19 years',]]
y=life['Life expectancy']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
lin_model=LinearRegression()
lin_model.fit(X_train,y_train)
y_test_predict=lin_model.predict(X_test)
rmse=np.sqrt(mean_squared_error(y_test,y_test_predict))
print('RMSE =',rmse)
plt.scatter(y_test,y_test_predict)
plt.plot(y_test_predict,y_test_predict,color='red',linestyle='--')
plt.show()
```
